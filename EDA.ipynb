    <p>Variables to cover:  NCP, CH2O, FAF, TUE

```python
og_obesity.columns
```

```python
from scipy import stats

sample1 = og_obesity['NCP']
sample2 = train_dlm['NCP'].sample(2111)


t_statistic_ncp, p_value_ncp = stats.ttest_ind(sample1, sample2)

print("t-statistic:", t_statistic_ncp)
print("p-value:", p_value_ncp)

alpha = 0.05  # significance level
if p_value_ncp < alpha:
    print("Reject null hypothesis: There is a significant difference between the means on NCP.")
else:
    print("Fail to reject null hypothesis: There is no significant difference between the means of NCP.")
```

```python
sample3 = og_obesity['CH2O']
sample4 = train_dlm['CH2O'].sample(2111)


t_statistic_ch20, p_value_ch20 = stats.ttest_ind(sample3, sample4)

print("t-statistic:", t_statistic_ch20)
print("p-value:", p_value_ch20)

alpha = 0.05  # significance level
if p_value_ch20 < alpha:
    print("Reject null hypothesis: There is a significant difference between the means of CH20.")
else:
    print("Fail to reject null hypothesis: There is no significant difference between the means of CH20.")
```

```python
sample5 = og_obesity['FAF']
sample6 = train_dlm['FAF'].sample(2111)


t_statistic_faf, p_value_faf = stats.ttest_ind(sample5, sample6)

print("t-statistic:", t_statistic_faf)
print("p-value:", p_value_faf)

alpha = 0.05  # significance level
if p_value_faf < alpha:
    print("Reject null hypothesis: There is a significant difference between the means of FAF.")
else:
    print("Fail to reject null hypothesis: There is no significant difference between the means of FAF.")
```

```python
sample7 = og_obesity['TUE']
sample8 = train_dlm['TUE'].sample(2111)


t_statistic_tue, p_value_tue = stats.ttest_ind(sample7, sample8)

print("t-statistic:", t_statistic_tue)
print("p-value:", p_value_tue)

alpha = 0.05  # significance level
if p_value_tue < alpha:
    print("Reject null hypothesis: There is a significant difference between the means of TUE.")
else:
    print("Fail to reject null hypothesis: There is no significant difference between the means of TUE.")
```
